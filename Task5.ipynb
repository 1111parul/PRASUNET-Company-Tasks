```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned dataset
df = pd.read_csv('RTA_Dataset_Cleaned.csv')

# Convert the 'Time' column to datetime and extract hour
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S', errors='coerce')
df['Hour'] = df['Time'].dt.hour

# Plot the distribution of accidents by hour of the day
plt.figure(figsize=(14, 8))
sns.countplot(data=df, x='Hour', order=df['Hour'].value_counts().index)
plt.title('Accidents by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Accidents')
plt.show()

# Plot the distribution of accidents by road surface conditions
plt.figure(figsize=(14, 8))
sns.countplot(data=df, x='Road_surface_conditions', order=df['Road_surface_conditions'].value_counts().index)
plt.title('Accidents by Road Surface Conditions')
plt.xlabel('Road Surface Conditions')
plt.ylabel('Number of Accidents')
plt.show()

# Plot the distribution of accidents by weather conditions
plt.figure(figsize=(14, 8))
sns.countplot(data=df, x='Weather_conditions', order=df['Weather_conditions'].value_counts().index)
plt.title('Accidents by Weather Conditions')
plt.xlabel('Weather Conditions')
plt.ylabel('Number of Accidents')
plt.show()

# Plot the relationship between road conditions and accident severity
plt.figure(figsize=(14, 8))
sns.countplot(data=df, x='Road_surface_conditions', hue='Accident_severity')
plt.title('Accident Severity by Road Surface Conditions')
plt.xlabel('Road Surface Conditions')
plt.ylabel('Number of Accidents')
plt.show()

# Plot the relationship between weather conditions and accident severity
plt.figure(figsize=(14, 8))
sns.countplot(data=df, x='Weather_conditions', hue='Accident_severity')
plt.title('Accident Severity by Weather Conditions')
plt.xlabel('Weather Conditions')
plt.ylabel('Number of Accidents')
plt.show()

# Plot the relationship between hour of the day and accident severity
plt.figure(figsize=(14, 8))
sns.countplot(data=df, x='Hour', hue='Accident_severity')
plt.title('Accident Severity by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Accidents')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned dataset
df = pd.read_csv('RTA_Dataset_Cleaned.csv')

# Define a list of columns to plot against accident severity
columns_to_plot = [
    'Driving_experience', 'Type_of_vehicle', 'Owner_of_vehicle',
    'Service_year_of_vehicle', 'Area_accident_occured', 'Lanes_or_Medians',
    'Road_allignment', 'Types_of_Junction', 'Road_surface_type',
    'Road_surface_conditions', 'Light_conditions'
]

# Plot graphs of these columns against severity of accidents
for column in columns_to_plot:
    plt.figure(figsize=(14, 8))
    sns.countplot(data=df, x=column, hue='Accident_severity', order=df[column].value_counts().index)
    plt.title(f'Accident Severity by {column.replace("_", " ").title()}')
    plt.xlabel(column.replace("_", " ").title())
    plt.ylabel('Number of Accidents')
    plt.xticks(rotation=45)
    plt.legend(title='Accident Severity')
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the cleaned dataset
df = pd.read_csv('RTA_Dataset_Cleaned.csv')

# Define a list of columns to plot against accident severity
columns_to_plot = [
    , 'Number_of_vehicles_involved', 'Number_of_casualties',
    'Sex_of_casualty', 'Age_band_of_casualty',
    'Casualty_severity',
]

# Plot pie charts of these columns against severity of accidents
for column in columns_to_plot:
    categories = df[column].value_counts().index
    num_categories = len(categories)
    num_cols = 2  # Number of columns for subplots
    num_rows = (num_categories + 1) // num_cols  # Calculate number of rows needed
    plt.figure(figsize=(20, 10))

    for i, category in enumerate(categories):
        plt.subplot(num_rows, num_cols, i + 1)
        severity_counts = df[df[column] == category]['Accident_severity'].value_counts()
        plt.pie(severity_counts, labels=severity_counts.index, autopct='%1.1f%%')
        plt.title(f'{column.replace("_", " ").title()} - {category}')

    # Adjust layout and add main title
    plt.suptitle(f'Accident Severity by {column.replace("_", " ").title()}', fontsize=16)
    plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.4)
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the cleaned dataset
df = pd.read_csv('RTA_Dataset_Cleaned.csv')

# Define the columns to analyze
columns_to_plot = [
    'Lanes_or_Medians', 'Road_allignment', 'Types_of_Junction',
    'Road_surface_type', 'Road_surface_conditions'
]

# Plot pie charts to mark accident hotspots
for column in columns_to_plot:
    categories = df[column].value_counts().index
    plt.figure(figsize=(20, 10))

    for i, category in enumerate(categories):
        plt.subplot(2, (len(categories) + 1) // 2, i + 1)
        severity_counts = df[df[column] == category]['Accident_severity'].value_counts()
        plt.pie(severity_counts, labels=severity_counts.index, autopct='%1.1f%%')
        plt.title(f'{column.replace("_", " ").title()} - {category}')

    plt.suptitle(f'Accident Severity by {column.replace("_", " ").title()} (Potential Hotspots)', fontsize=16)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
```

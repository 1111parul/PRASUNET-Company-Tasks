```python
# import pandas as pd
# from google.colab import files
# import matplotlib.pyplot as plt
# import seaborn as sns

# # Upload the file
# uploaded = files.upload()

# Load the dataset (assuming the file is named 'social_media_data.csv')
df = pd.read_csv('twitter_training-1 - Sheet1 (1).csv', delimiter=',')

# Inspect the first few rows of the dataset to verify correct loading
print(df.head())

# Data Preprocessing
df.dropna(subset=['Sentiment'], inplace=True)

# Verify the unique values in the 'Sentiment' column
print(df['Sentiment'].unique())

# Plot sentiment distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='Sentiment', data=df, palette='viridis')
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
# Plot a pie chart for the Sentiment column
sentiment_counts = df['Sentiment'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis', len(sentiment_counts)))
plt.title('Sentiment Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
